# Two Variable plot
Adapted from [https://matplotlib.org/tutorials/introductory/pyplot.html](https://matplotlib.org/tutorials/introductory/pyplot.html).










```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# linspace takes the start value the end value and the amount
# of points inbetween those values you want
x =np.linspace(0.0, 100, 101)
y1 = x*x
y2 = x*x*x
y3 = x*x*x*x

plt.plot(x,y1,'g')
plt.plot(x,y2,'r')
plt.plot(x,y3,'b')
```

```python
x =np.linspace(0.0, 100, 101)
y1 = 2.0 * x + 5.0
y2 = 1.0 * x - 5.0
y3 = -2.0 * x + 0.0

plt.plot(x,y1,'g')
plt.plot(x,y2,'r')
plt.plot(x,y3,'b')
```

```python
# Passing in a string with the plot function which will determine
# colour and the type of plot that will be displayed
plt.plot([1,2,3,4], [1,4,9,16], "gx")
```

```python
# if you only give plt one array of numbers it will assume they are the 
# y-axis numbers
# When given two arrays it can plot both the x and y axis numbers
# on the graph
plt.plot([1,2,3,4], [1,4,9,16])
```
